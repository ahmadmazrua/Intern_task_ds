##  importing required packages 

```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Reading a csv file into Dataframe and show the first 5 rows

```python
df = pd.read_csv('drinkMenu.csv')
df.head()
```

## summarize the central tendency, dispersion and shape of a dataset’s distribution.

```python
df.describe()
```

## Remove duplicate rows 

```python
df.drop_duplicates(keep=False,inplace=True)
```

## Fill null values 

```python
df.isnull().sum()
```

## Row containing null value

```python
df[df['Caffeine (mg)'].isnull()]
```

```python
df[df['Beverage_category'] == 'Shaken Iced Beverages']
```

## Choose the columns of interest

```python
df_col_of_intr = df[['Beverage','Calories',' Sugars (g)']]
df_col_of_intr
```

# Visulaization 

## Sorting Beverage by Calories to find Highest value

```python
df_h_c=df_col_of_intr.sort_values(by=['Calories'], ascending=False)
calories = df_h_c[['Beverage', 'Calories']][:10]
calories
```

## Ploting Beverage with Calories that showing highest calories

```python
fig, ax = plt.subplots()
ax.bar(range(len(calories['Beverage'])),calories['Calories'])
ax.set_xticks(range(len(calories['Beverage'])))
ax.set_xticklabels(calories['Beverage'].values)
plt.xticks(rotation = 90)
_ = plt.show()
```

```python

calories['Beverage'].values
```

## Sorting Beverage by Sugars to find Highest value

```python
df_h_s=df_col_of_intr.sort_values(by=[' Sugars (g)'], ascending=False)
Sugar = df_h_s[['Beverage', ' Sugars (g)']][:10]
Sugar
```

## Ploting Beverage with Sugars that showing highest Sugars

```python
fig, ax = plt.subplots()
ax.bar(range(len(Sugar['Beverage'])),Sugar[' Sugars (g)'])
ax.set_xticks(range(len(Sugar['Beverage'])))
ax.set_xticklabels(Sugar['Beverage'].values)
plt.xticks(rotation = 90)
_ = plt.show()
```
